```python
import bs4
import requests
```

```python
req = requests.get("https://fr.wikipedia.org/wiki/Liste_des_communes_de_France_les_plus_peuplées")
print(req)
```

```python
soup = bs4.BeautifulSoup(req.text)
soup.find('table', class_ = "wikitable sortable jquery-tablesorter")

print(soup.findAll("td", style = "text-align:left"))
```

```python
soup = bs4.BeautifulSoup(req.text)

L=[]
n=0
for x in soup.findAll("td", style = "text-align:left"):
    if n == 30:
        break
    elif n%3 == 0:
        L.append(x.text.replace("\n",""))
    n+=1
    
print(L)
```
