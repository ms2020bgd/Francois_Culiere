```python
import bs4
import requests
import numpy as np
import pandas as pd
```

```python
def get_most_populated_cities(soup):
    L=[]
    n=0
    for x in soup.findAll("td", style = "text-align:left"):
        if n == 30:
            break
        elif n%3 == 0:
            L.append(x.text.replace("\n","").replace('''[19]''', ""))
        n+=1

    return L
```

```python
def get_distance(origin, destination):
    url = f"https://fr.distance24.org/route.json?stops={origin}|{destination}"
    data = requests.get(url).json()
    return data['distance']
```

```python
req = requests.get("https://fr.wikipedia.org/wiki/Liste_des_communes_de_France_les_plus_peupl√©es")
soup = bs4.BeautifulSoup(req.text)
cities = get_most_populated_cities(soup)
M = np.zeros((10,10))

for i in range(len(cities)):
    for j in range(len(cities)):
        M[i,j] = get_distance(cities[i], cities[j])  
        
df = pd.DataFrame(M, index=cities, columns=cities)
df
```
